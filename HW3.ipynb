```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

# Q1

```python
gold = pd.read_table("gold.txt",header = None,names=["url","category"])
labels=pd.read_table("labels.txt",header = None,names=["turks","url","category"])
labels
```

# Q2

```python
labels_on_gold = labels[labels["url"].isin(gold["url"])]
labels_unknown = labels[~labels["url"].isin(gold["url"])]
labels_on_gold 
```

# Q3

```python
new_data = labels.merge(gold,left_on ="url" ,right_on = "url",suffixes=["_labels","_gold"])
correct_num_turks = (new_data[new_data["category_labels"] == new_data["category_gold"]]["turks"].value_counts())
Total_num_per_turks = new_data["turks"].value_counts()
average_correctness = correct_num_turks/Total_num_per_turks
average_correctness = average_correctness.fillna(0)
rater_goodness = DataFrame(dict(Total_num_per_turks=Total_num_per_turks,average_correctness=average_correctness))
rater_goodness
```

# Q4

```python
rater_goodness["odds"]=rater_goodness['average_correctness']/(1.001-rater_goodness['average_correctness'])
rater_goodness
```

# Q5

```python
rated_high = rater_goodness[rater_goodness["Total_num_per_turks"]>=20]
rated_high.sort_values(by='average_correctness',ascending=False)[:10]
```

# Q6

```python
plot(rater_goodness["Total_num_per_turks"],rater_goodness["average_correctness"],marker='o',color='blue', linestyle='None')
xlabel("Total_num_per_turks")
ylabel("average_correctness")
legend(numpoints=1, loc='best')
```

# Q7

```python
reliable_turks = rater_goodness[rater_goodness["Total_num_per_turks"]>rater_goodness["Total_num_per_turks"].quantile(0.75)]
append_odds = reliable_turks[[ 'odds']].merge(labels_unknown,left_index=True,right_on="turks")
#append_odds.groupby('turks')["turks"].agg('count')
grps = append_odds.groupby(["url","category"]).prod()
grps
```

# Q8

```python
result_75=pd.DataFrame()
grps_categ = grps.unstack("category")
top_odds = grps_categ.T.max()
result_75['top odds'] = top_odds
result_75["top category"]=grps_categ["odds"].T.idxmax()
result_75
```
